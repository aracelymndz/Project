```python
# Dependencies and Setup
%matplotlib inline
import matplotlib.patches as mpatches
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import json
import datetime 
from pandas import Timestamp


file_to_load = "parking.csv"


#Read Purchasing File and store into Pandas data frame
data_tickets = pd.read_csv(file_to_load)
data_tickets.head(2)
```

```python
# print(data_tickets['Issue Date'].isna().sum())

# (Basically convert the NA/NAn on ceros at the column Issue Data )
data_tickets['Issue Date'].fillna(0, inplace=True)

#convert str to dtype
data_tickets['Issue Date'] = data_tickets['Issue Date'].astype(str)

#Get all items star with 2018 at Issu Date 
data_tickets_2018 = data_tickets[data_tickets['Issue Date'].str.contains("2018")]


data_tickets_2018.sample(5)
```

```python
# count unique values 
violation_count = data_tickets_2018["Violation code"].value_counts()

#Get the 10 tops 
violation_count.nlargest(10)
```

```python
bins = [0, 59, 159, 259, 359, 459, 559, 659, 759, 859, 959, 1059, 1159, 1259, 1359, 1459, 1559, 1659, 1759, 1859, 1959, 2059, 2159, 2259, 2359]
group_names = ['12am-1am', '1am-2am', '2am-3am', '3am-4am', '4am-5am', '5am-6am', '6am-7am', '7am-8am', '8am-9am', '9am-10am', '10am-11am', '11am-12pm', '12pm-1pm', '1pm-2pm', '2pm-3pm', '3pm-4pm', '4pm-5pm', '5pm-6pm', '6pm-7pm', '7pm-8pm', '8pm-9pm', '9pm-10pm', '10pm,11pm', '11pm-12am']
data_tickets_2018["Time Day"] = pd.cut(data_tickets_2018["Issue time"], bins, labels=group_names)

hrs_group_demo = data_tickets_2018["Time Day"].value_counts()
tickets_x_hrs = data_tickets_2018.groupby("Time Day")
total_viola_cout = tickets_x_hrs ["Violation code"].value_counts()
# total_viola_cout


df = pd.DataFrame(hrs_group_demo)

```
